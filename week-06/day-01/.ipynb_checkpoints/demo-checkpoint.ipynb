```python
# What are the ten most recently updated apps?

import calendar
def format_transfer(string_datetime):
    if ".0." not in string_datetime:
        tmp = string_datetime.split(",")
        year = tmp[1]
        month_day = tmp[0]
        month = month_day.split(' ')[0]
        month = list(calendar.month_name).index(month)
        day = month_day.split(' ')[1]
        return "-".join([str(year),str(month),str(day)])
    
store['date_time'] = store['Last Updated'].apply(format_transfer)
store['date_time'] = pd.to_datetime(store['date_time'])
sorted_value = store.sort_values(by="date_time",inplace=False,ascending=False)
#store.sort_values(by = ['date_time'])
sorted_value
```

```python
# Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.
from datetime import date
month = sorted_value.date_time.max().month - 2
year = sorted_value.date_time.max().year
day = sorted_value.date_time.max().day
tmp = str(year) + str(month) + str(day)
line = pd.to_datetime(tmp,format='%Y%m%d')

top_25 = store.sort_values(by = "Rating",ascending = False)[:26]
last_two_month = store[store['date_time'] > line]
last_two_month
line


innersect = pd.merge(top_25,last_two_month, how='inner')
innersect

```
