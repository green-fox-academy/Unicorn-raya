```python
import sklearn
import pandas as pd
```

```python
abspath_training_data = "C:\\Users\\Ray_Zhang\\greenfox\\Unicorn-raya\\week-06\\day-04_05\\titanic\\train.csv"
abspath_test_data = "C:\\Users\\Ray_Zhang\\greenfox\\Unicorn-raya\\week-06\\day-04_05\\titanic\\test.csv"
abspath_test_data_result = "C:\\Users\\Ray_Zhang\\greenfox\\Unicorn-raya\\week-06\\day-04_05\\titanic\\gender_submission.csv"

train_data = pd.read_csv(abspath_training_data)
test_data = pd.read_csv(abspath_test_data)
test_data_label = pd.read_csv(abspath_test_data_result)
```

```python
train_data.info()
print("-"*30)
test_data.info()
print("-"*30)
test_data_label.info()
```

```python
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)

test_data['Age'].fillna(test_data['Age'].mean(),inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(),inplace=True)
# a = train_data[ train_data['Age'] == 'null']
```

```python
print(train_data['Embarked'].value_counts())
a = train_data[ train_data['Embarked'].isnull()]
a
```

```python
train_data['Embarked'].fillna('S', inplace=True)
test_data['Embarked'].fillna('S',inplace=True)
a = train_data[ train_data['Embarked'].isnull()]
a
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
train_features = train_data[features]
train_labels = train_data['Survived']
test_features = test_data[features]
test_labels = test_data_label['Survived']
```

```python
train_features
test_features
```

```python
from sklearn.feature_extraction import DictVectorizer
dvec=DictVectorizer(sparse=False)
train_features=dvec.fit_transform(train_features.to_dict(orient='record'))
```

```python
print(dvec.feature_names_)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()#criterion='entropy'
clf.fit(train_features, train_labels)
```

```python
test_features=dvec.transform(test_features.to_dict(orient='record'))
pred_labels = clf.predict(test_features)
```

```python
acc_traindata = round(clf.score(train_features, train_labels), 6)
print(f"training accuracy: {acc_traindata}")
acc_testdata= round(clf.score(test_features, test_labels), 6)
print(f"testing accuracy: {acc_testdata}")
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
print(f'cross_val_score: { round(np.mean(cross_val_score(clf, train_features, train_labels, cv=10)),4)}')
```
